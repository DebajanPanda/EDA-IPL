```python
batting_first=ipldata[ipldata['win_by_runs']!=0]
batting_first.head()
# credit goes to bowler


```

```python
plt.figure(figsize=(15,4))
plt.bar(batting_first['win_by_runs'].value_counts().index,batting_first['win_by_runs'].value_counts().values)
plt.title('win_by_runs')
plt.xlabel('runs')
plt.ylabel('count')
plt.show()
#strong bowling team
```
